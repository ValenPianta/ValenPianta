# Ejemplos

Graficar las funciones $f(x) = -0.2x+1$ y $g(x)=ln(x)/2$.

```python
valores_de_x = np.arange(0.2, 10, 0.5)  # Esta función es como range, pero para floats
valores_de_f = -0.2*valores_de_x + 1    # Los arrays de numpy permiten hacer operaciones matemáticas
valores_de_g = np.log(valores_de_x)/2

plt.plot(valores_de_x, valores_de_f, label='f(x)')
plt.plot(valores_de_x, valores_de_g, label='g(x)')
plt.legend()    # Esto hace que se muestre la cajita con los nombres de las funciones
plt.xlabel('x') # Acá le podés poner la descripción que quieras a los ejes
plt.ylabel('y')
plt.title('Dos funciones')
plt.show()
```

# Tiempo de descarga

El tiempo que tarda en descargarse un archivo no es siempre igual y no se puede predecir de manera exacta. Por lo tanto, es una variable aleatoria.

La función `descargar_y_medir_tiempo` descarga un archivo de internet y devuelve el tiempo que demoró la descarga, en segundos.

```python
from urllib.request import urlretrieve
from timeit import timeit

def descargar():
    urlretrieve('https://tinyurl.com/archivomisterioso')

def descargar_y_medir_tiempo():
    return timeit(descargar, number = 1)
```

```python
# Cada vez que ejecutás esto, da un número distinto:
print(f'La descarga demoró {descargar_y_medir_tiempo()} segundos.')
```
